Function to calculate de the normalized logarithmic returns 

```python
def r_log(price):
    price = price[price > 0]
    P1 = np.roll(price,-1)
    R = np.log(P1)-np.log(price); R = R[:-1]
    r = (R-np.mean(R))/np.std(R)
    return r
    
    
```
